```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, MinMaxScaler

# Drop the 'Route' column
if 'Route' in dt. columns:
        dt.drop(columns=["Route"],inplace=True)

# Impute missing values in 'Total_Stops' with the most frequent value
most_frequent_value = dt['Total_Stops'].mode()
dt['Total_Stops'].fillna(most_frequent_value, inplace=True)

# Define ordinal encoding orders
total_stops_order = ['non-stop', '1 stop', '2 stops', '3 stops', '4 stops']
time_order = ['Morning', 'Afternoon', 'Evening', 'Night']

# Create separate encoders
default_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
ordered_encoder = OrdinalEncoder(categories=[total_stops_order, time_order, time_order],
                                 handle_unknown='use_encoded_value', unknown_value=-1)

# Features to encode
default_features = ['Airline', 'Source', 'Destination', 'Additional_Info']
ordered_features = ['Total_Stops', 'Dep_Time', 'Arrival_Time']

# Fit and transform
dt[default_features] = default_encoder.fit_transform(dt[default_features])
dt[ordered_features] = ordered_encoder.fit_transform(dt[ordered_features])

# Function to convert duration to total minutes
def convert_duration_to_minutes(duration):
    parts = duration.split()
    total_minutes = 0
    for part in parts:
        if 'h' in part:
            total_minutes += int(part.replace('h', '')) * 60
        elif 'm' in part:
            total_minutes += int(part.replace('m', ''))
    return total_minutes

# Convert and scale 'Duration'
dt['Duration'] = dt['Duration'].apply(convert_duration_to_minutes)
scaler = MinMaxScaler()
dt[['Duration']] = scaler.fit_transform(dt[['Duration']])

# Check for unseen values in test dataset
unseen_values = (dt[default_features + ordered_features] == -1).any().any()

# Output result for Q10
q10_answer = "Yes" if unseen_values else "No"
print("Answer to Q10:", q10_answer)
```
