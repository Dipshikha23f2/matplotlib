```python
# from a pandas dataframe
batsman=pd.read_csv("/content/sharma-kohli.csv")
batsman
plt.plot(batsman['index'],batsman['V Kohli'],color='#fbff33') # use hex code on crome
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ff3333') # use hex code on crome for color
plt.title(' Rohit vs Virat career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
#color(hex code) & line(width & style) & marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#fbff33') #linestyle - (change the style of line-solid/dt/dashdot/dash)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ff3333') #linewidth(we can change width of line like 2,3...etc)
plt.title(' Rohit vs Virat career comparison')#marker(we change marker at tilled point like(diamond,+,dot etc ,ex-(marker='D))
plt.xlabel('season') #(markersize=(size of marker(like 2,4..)))
plt.ylabel('runs scored')
```

```python
# line(width & style) & marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#fbff33',linestyle='solid',linewidth=3,marker='+',markersize=20)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ff3333',linestyle='dashdot',linewidth=2,marker='D',markersize=7)
plt.title(' Rohit vs Virat career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#fbff33',linestyle='solid',linewidth=3,marker='+',markersize=20,label='Rohit')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ff3333',linestyle='dashdot',linewidth=2,marker='D',markersize=7,label='Virat')
plt.title(' Rohit vs Virat career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
plt.legend(loc='upper left')
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Define the tasks and their respective start dates and durations
tasks = [
    ("Data Collection", "2025-01-20", 16),  # duration in days
    ("Data Cleaning & Exploration", "2025-02-05", 26),
    ("Data Analysis", "2025-03-03", 12),
    ("Implementation Segment Analysis", "2025-03-15", 10),
    ("Integration of Analytics Finding", "2025-03-25", 15),
    ("Report Compilation & Recommendation", "2025-04-10", 10),
]

# Convert start dates to datetime objects
task_names = [task[0] for task in tasks]
start_dates = [datetime.strptime(task[1], "%Y-%m-%d") for task in tasks]
durations = [task[2] for task in tasks]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the Gantt chart
for i, (task, start, duration) in enumerate(zip(task_names, start_dates, durations)):
    ax.barh(task, duration, left=start, color="red", edgecolor="black")

# Format the x-axis with date labels
ax.xaxis.set_major_formatter(mdates.DateFormatter("%d-%b-%Y"))
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))  # Show ticks every 2 weeks
plt.xticks(rotation=45)

# Add labels and title
plt.xlabel("Timeline")
plt.ylabel("Tasks")
plt.title("Project Timeline - Gantt Chart")
plt.tight_layout()

# Show the plot
plt.show()
```
