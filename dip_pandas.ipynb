## Pandas Series
A Pandas Series is like a column in a table. It is a 1-D array holding data of any type

#Importing pandas

```python
import numpy as np
import pandas as pd
```

## Series from the list

```python
# string
country = ['india','pakistan','usa','nepal','Srilanka']
pd.Series(country)
```

```python
# integers
runs = [13,24,56,78,100]
runs_ser = pd.Series(runs)
print(runs_ser)
```

```python
# custom index
marks=[67,78,90,54]
sub=['sci','hindi','eng','math']
pd.Series(marks,index=sub)
```

```python
# setting a name
marks = pd.Series(marks,index=sub,name='dip_marks')
marks
```

```python
marks = {'maths':67,'english':57,'science':89,'hindi':100}
marks_series = pd.Series(marks,name='deep ke marks')
marks_series
```

## Series Attributes
